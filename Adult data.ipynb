```python
import pandas as pd

columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation',
 'relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country', 'Earnings-Raw']

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datamining_project/adult.data')

df.columns = columns
# data 불러오기
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing

enc = OneHotEncoder(sparse = False)

hot = ['sex', 'Earnings-Raw']
for item in hot:
    out_enc = enc.fit_transform(df[[item]])
    df[item] = out_enc
# One hot encoding을 이용한 전처리

enc_1 = preprocessing.LabelEncoder()

label = ['workclass','education','marital-status','occupation','relationship','race','native-country']
for item in label:
    out_enc_1 = enc_1.fit_transform(df[item])
    df[item] = out_enc_1
 # Label encoding을 이용한 전처리
```

```python
cor = df.corr()

cor_target = abs(cor['Earnings-Raw'])

selected_col = cor_target[cor_target>0.22]

df_new = df.loc[:, selected_col.index]
# cor을 이용하여 상관계수를 계산후 상관계수가 0.22 초과인 column 선택
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score

X_train, X_test, y_train, y_test = train_test_split(df_new.loc[:, df_new.columns!='Earnings-Raw'], df_new['Earnings-Raw'], test_size = .33, random_state=42)
# 선택한 열로 train set, test set을 만듦

params = {"max_depth":[2,3,4,5]}

tree = DecisionTreeClassifier(random_state = 42)
clf_grid_tree = GridSearchCV(tree, params, cv=5, scoring='f1_macro')
clf_grid_tree.fit(X_train, y_train)

y_pred_tree = clf_grid_tree.predict(X_test)
f1_tree = f1_score(y_test, y_pred_tree)
# DecisionTreeClassifier를 이용한 GridSearch

print(f1_tree)
```
