```python
import pandas as pd
import numpy as np

basket_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datamining_project/basketball.csv')
basket_data.index.name = "index"
# data 불러오기
```

# 1-1 해당 시즌 우승팀 찾기

```python
basket_data["Winner"] = 0

for i in range(len(basket_data)):
  if basket_data["VisitorPts"][i] > basket_data["HomePts"][i]:
    basket_data["Winner"][i] = basket_data["Visitor Team"][i]
  else:
    basket_data["Winner"][i] = basket_data["Home Team"][i]
# Winner라는 column을 생성하여 그 날 경기에서 이긴 팀을 입력
```

```python
Team = pd.DataFrame(pd.unique(basket_data["Winner"]), columns = ['Team name'])
Team["Win count"] = 0
# 팀 별 이긴 횟수 세기 위한 새로운 dataframe 및 row 생성

for i in range(len(basket_data)):
  for j in range(len(Team)):
    if basket_data["Winner"][i] == Team["Team name"][j]:
      Team["Win count"][j] += 1
# basket_data를 순회하며 Winner와 이름이 같은 Teams에 Win count를 1씩 증가
```

```python
for i in range(len(Team)):
    if Team["Win count"][i] == max(Team["Win count"]):
        print(Team["Team name"][i])
# Win count가 최대인 열의 팀 이름과 같은 팀의 이름 출력
```

# 1-2 홈팀의 승패를 예측하는 모델만들기

```python
basket_data["Home team win"] = 0

for i in range(len(basket_data)):
  if basket_data["Winner"][i] == basket_data["Home Team"][i]:
      basket_data["Home team win"][i] = 1
# 홈팀이 이겼는지 여부를 'Home team win' column에 입력
```

```python
# 관중이 많으면 홈팀이 승리한다고 예측
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
train_test_split(basket_data.loc[:, basket_data.columns == 'Attend.'], basket_data["Home team win"], test_size=.33, random_state=42)
# 관중수 data를 이용하여 홈팀의 승리여부를 예측하기 위한 test set과 train set을 만듦
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score

clf_tree = DecisionTreeClassifier(max_depth=3, random_state = 42)
clf_tree.fit(X_test, y_test)

y_pred_tree = clf_tree.predict(X_test)
f1_tree = f1_score(y_test, y_pred_tree)
# DecisionTreeClassifier를 이용하여 관중수에 따른 승리 여부를 예측함

print(f1_tree)
```
